```python
import random
import time

print("""
      
      Choose Your Cup! Game
      
      Please choose one of the following: -
      
      1 = Cup 1
      2 = Cup 2
      3 = Cup 3
      4 = Exit
      
      Rules
      
      1. If you guess which cup the ball is under, you win!
      2. You don't guess which cup the ball is under, you lose!
      
      Have fun!
      
      """)


while True:

    User_1 = int(input("Choose your cup: "))
    
    if User_1 == 4:
        print("You have chosen to exit. Thank you for playing!")
        break
    
    elif User_1 >= 5:
        print("Please choose Cup 1, Cup 2, Cup 3 only (or 4 to exit).")
        
    else:
        
        Game = {1:"Cup 1", 2:"Cup 2", 3:"Cup 3"}
        
        AI_1 = random.randint(1,3)
        
        AI = Game.get(AI_1)
        User = Game.get(User_1)
        
        if User == "Cup 1" and AI == "Cup 1":
            time.sleep(0.1)
            print("The ball was under Cup 1. You win!")
            
        elif User == "Cup 1" and AI == "Cup 2":
            time.sleep(0.1)
            print("The ball was under Cup 2. You lose!")
            
        elif User == "Cup 1" and AI == "Cup 3":
            time.sleep(0.1)
            print("The ball was under Cup 2. You lose!")
            
        elif User == "Cup 2" and AI == "Cup 1":
            time.sleep(0.1)
            print("The ball was under Cup 1. You lose!")
            
        elif User == "Cup 2" and AI == "Cup 2":
            time.sleep(0.1)
            print("The ball was under Cup 2. You win!")
            
        elif User == "Cup 2" and AI == "Cup 3":
            time.sleep(0.1)
            print("The ball was under Cup 2. You lose!")
        
        elif User == "Cup 3" and AI == "Cup 1":
            time.sleep(0.1)
            print("The ball was under Cup 1. You lose!")
            
        elif User == "Cup 3" and AI == "Cup 2":
            time.sleep(0.1)
            print("The ball was under Cup 2. You lose!")
            
        elif User == "Cup 3" and AI == "Cup 3":
                time.sleep(0.1)
                print("The ball was under Cup 3. You win!")
            
        else:
            print("Please only choose numbers 1, 2, 3 or 4.")
```
